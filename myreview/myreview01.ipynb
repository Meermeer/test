결측치

```python
import pandas as pd
import numpy as np
```

```python
_value = [
    [ 1,2,3 ],
    [ 4,np.nan,6 ],     # np.nan : 결측치를 생성한다
    [ 7,8,np.nan ]
]
df = pd.DataFrame(_value)
df
```

```python
# 결측치를 확인
# 데이터 프레임의 정보를 확인
df.info()
```

```python
# is() 함수를 사용하여 결측치를 확인
# isna() 함수 사용. isna() : 결측치면 True, 결측치가 아니면 False를 출력
df.isna().sum()     # 첫 번째 인덱스에는 결측치가 0개, 두 번째랑 세 번째 인덱스에는 결측치가 1개
```

```python
df
```

```python
# 결측치를 제거
# dropna()는 결측치를 제거 하는 함수
# 매개변수 axis : 0,

# df.dropna( axis=1 ) 또는 df.dropna( axis='columns') : 결측치가 있는 컬럼(열)을 제거(세로로 자르는 느낌)
df.dropna( axis='columns')
```

```python
# df.dropna( axis=0 ) 또는 df.dropna( axis='rows' ) : 결측치가 있는 행을 제거(가로로 자르는 느낌)
df.dropna( axis='rows' )
```

```python
# sales 데이터 로드
sales = pd.read_csv( "../csv/Sales Records.csv" )
sales.head(2)
```

```python
sales.sort_values( 'Region', ascending=True )     # 'Region' 컬럼의 값들(values)을 오름차순(작은거부터 큰거)으로 배열한다(Default가 오름차순). 
                                                  # inplace = True : 해당하는 작업을 진행 후 해당 변수에 저장. 변화를 반형한다는 의미
```

```python
sales.sort_values( 'Region', ascending=False )  # 'Region' 컬럼의 값들(values)을 'ascending=False'시 내림차순(큰거부터 작은거)으로 배열한다.
```

```python
sales.sort_values( 'Region', inplace=True )     # 'Region' 컬럼의 값들(values)을 오름차순(작은거부터 큰거)으로 배열후 그 변화를 데이터프레임 sales에 반영한다. 
```

```python
sales
```

```python
flg = sales[ 'Region' ] == 'Asia'   #'Region' 컬럼의 값들을 각각 비교하여 Asia가 맞으면 True를 아니면 False를 출력
flg
```

```python
# 데이터 프레임 필터 기능
# loc[ 행의 조건, 열의 조건 ]

# 행의 조건을 가지고 필터
sales.loc[~flg]     # 'Asia'가 포함되지 않은 행들만 출력한다
```

```python
# 열의 조건을 가지고 필터
sales.loc[ :, ['Region'] ]     # 'Region' 컬럼의 내용만 출력한다
```

```python
# 행과 열의 조건을 모두 적용시켜서 필터
sales.loc[ flg, ['Region'] ]       # 'Region' 컬럼의 'Asia'가 포함된 내용들만 출력
```

```python
# iloc[ 행의 위치, 열의 위치 ]
sales.iloc[ 1:3, 1:3 ]  #행의 두번째 인덱스부터 세번째 인덱스까지, 컬럼의 두번째 인덱스부터 세번째 인덱스까지의 내용만 출력
```

```python
sales.iloc[ 1:3 ]      # 두 번째 행부터 세 번째 행까지의 내용만 출력
```

```python
sales[ ['Region'] ]     # sales.loc[ :, ['Region'] ]과 동작이 같음, sales['Region']로 쓰면 시리즈 형태로 출력된다.
```

```python
df.isna().any(axis = 'columns')     # df.isna().any(axis = 1)
```

먼저 "df.isna()"는 "df" 데이터프레임 내의 모든 원소에 대해 결측값 여부를 True/False로 반환하는 데이터프레임을 생성한다. 
이 때, "isna()"는 각 원소가 결측값인지 아닌지를 판별하여, 결측값이면 True, 결측값이 아니면 False를 반환합니다.

그 다음, ".any(axis = 'columns')"는 행(row)을 기준으로 각 열(column)에 대한 True/False 값을 OR 연산하여 결과값을 반환합니다. 
즉, 각 행마다 하나 이상의 결측값이 있으면 True를 반환하고, 그렇지 않으면 False를 반환합니다.

따라서, "df.isna().any(axis = 'columns')" 코드는 "df" 데이터프레임 내의 각 행(row)에 대해, 하나 이상의 결측값이 있는지를 검사하여 True/False 값을 반환하는 코드입니다.

```python
df.isna().any(axis = 'rows')    # any(axis = 0) 혹은 any(axis = 'rows') : 각 행(rows)에대한 True/False값을 OR 연산하여 결과값을 반환 
```

```python
flg_na = df.isna().any(axis = 'columns')    # 결측치가 있는 하나라도 있는 행은 True, 결측치가 전혀 없는 행은 False
flg_na
```

```python
df
```

```python
df.loc[~flg_na]     # 결측치 값이 없는 행을 출력한다
```
