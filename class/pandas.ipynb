### 경로

    - 환경이 변하더라도 같은 위치를 지정

결측치

```python
import pandas as pd
import numpy as np
```

```python
_value = [
    [1,2,3],
    [4,np.nan,6],
    [7,8,np.nan]
]
df = pd.DataFrame(_value)
df
```

```python
# is() 함수를 사용하여 결측치를 확인
# isna() 함수 사용
df.isna().sum()
```

```python
# 결측치를 제거
# dropna()는 결측치를 제거 하는 함수
# 매개 변수 axis : 0 또는 'rows'을 대입하면 해당하는 행을 삭제
#                 1 또는 'columns'을 대입하면 해당하는 열알 삭제
df.dropna(axis = 1)
```

```python
# 데이터 프레임 필터 기능
# loc[행위 조건, 열의 조건]
df.loc[0]

# iloc[행의 위치, 열의 위치]
```

```python
# sales 데이터 로드
sales = pd.read_csv("../csv/Sales Records.csv")
#sales.head(2)
sales.loc[0]
```

```python
flg = sales['Region'] == 'Asia'
sales.loc[flg]
```

```python
# flg2 = sales['Region'] != 'Asia'
flg2 = sales['Region'] == 'Asia'

# 행의 조건을 가지고 필터
sales.loc[~flg2]   # ~ : True와 False를 뒤집음
```

```python
# 열의 조건을 가지고 필터
sales.loc[:, ['Order Priority', 'Order Date']]
```

```python
# 열의 조건을 가지고 필터
sales.loc[:, ['Region']]
```

```python
sales[['Region']]
```

```python
# 행과 열의 조건을 모두 적용시켜서 필터
sales.loc[flg,['Region']]
```

```python
# iloc[행의 위치, 열의 위치]
sales.iloc[1:3, 1:3]
```

```python
sales.loc[[1,2]]
```

```python
# 결측치가 포함된 데이터만 필터
df.columns = ['A', 'B', 'C']
fla_na = df.isna().any(axis=1)
df
```

```python
df.loc[fla_na]      # 결측치가 포함된 데이터만 뽑음
```

```python
df.loc[~fla_na]     # 결축츠가 아닌 데이터만 봅음
```

```python
uriage = pd.read_csv("../csv/uriage.csv")
uriage
```

```python
# 결측치를 확인
uriage.isna().sum()
```

```python
uriage.info()
```

```python
# 결측치가 존재하는 조건 생성
uriage.isna().any(1) # 행의 조건
```

```python
uriage.loc[uriage.isna().any(1)] # 행의 결측치인 조건을 걸고 열은 다 보여줌
```

```python
# 결측치가 아닌 값들만 출력
uriage.loc[~uriage.isna().any(1)] # 행의 결측치가 아닌 조건을 걸고 열은 다 보여줌
```

```python
uriage.dropna(axis=0)   # 결측치가 포함된 행을 다 날려 버린다
```
