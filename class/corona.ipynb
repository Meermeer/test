```python
import pandas as pd
import numpy as np
```

1. corona.csv 파일 로드
2. 쓸모없는 컬럼을 삭제
3. 컬럼의 이름을 변경
4. 특정한 컬럼의 값을 기준으로 정렬 변경
5. 인덱스를 초기화

```python
corona = pd.read_csv("../csv/corona.csv") # "Unnamed: 0"는 필요 없는 Column이다.
corona.head(2)
```

```python
# unnamed: 0 컬럼을 삭제
corona = corona.drop("Unnamed: 0", axis = 1)
```

```python
# 인덱스 0을 삭제
corona.drop(0,axis=0)
```

```python
corona.head(1)
```

```python
# 컬럼의 이름을 변경
# 컬럼 변수를 출력
corona.columns = ["등록일시", "누적사망자", "누적확진자", "게시글번호",
                  "기준일", "기준시간", "수정일시", "누적의심자", "누적확진률"]
```

```python
corona.head(1)
```

```python
corona.head(5) # 현재 등록일시가 내림차순으로 정렬되있음
```

```python
# 오름차순으로 바꾸기 위해 sort() 함수를 씀
# sort_values()
# 매개변수 ascending : True(기본값)(오름차순), False(내림차순)
# 매개변수 inplace : 기준이 되는 데이터 프레임을 수정할것인가를 지정

corona.sort_values('등록일시', ascending = True, inplace=True) # inplace로 인해 corona 데이터에 뭔가 변화가 있었음
```

```python
corona.head(5)
```

```python
# 인덱스를 초기화 
# reset_index()
# 매개변수 drop : 기본값(False) -> 기존의 인덱스의 값을 삭제할지 지정
# 매개변수 inplace : 기준이 되는 데이터프레임을 수정할지 지정
# corona.reset_index() : 기존의 인덱스가 column으로 들어가서 새로운 시리즈가 되버렸음

corona.reset_index(drop=True) # corona.reset_index().drop('index',axis=1)랑 동작이 같음

corona.reset_index(drop=True, inplace=True) #기존의 인덱스 값을 삭제한걸 반영함. corona = corona.reset_index(drop=True)랑 같음
```

```python
corona.head(5)
```

```python
corona
```

```python
# 데이터간의 차이를 출력하는 함수
# diff(n)
# n의 기본값은 1 -> 1전 인덱스 데이터를 선택
# corona.누적확진자.diff(1) 
#corona['누적확진자']

corona["일일확진자"] = corona.누적확진자.diff(1) 
```

```python
corona.tail(5)
```

```python
# shift() : 이동한 값을 출력
corona2 = corona.head(5)
```

```python
corona2['누적사망자']
```

```python
corona2['누적사망자'].shift(1)
```

```python
# 일일 사망자 파생변수를 생성
corona['일일사망자'] = corona['누적사망자'] - corona['누적사망자'].shift(1)
```

```python
corona2['누적사망자'] - corona2['누적사망자'].shift(1)
```
