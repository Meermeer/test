### 연습문제
1. csv 폴더에 있는 Sales Records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 삭제
3. 컬럼의 이름을 변경. ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 권역과 국가를 기준으로 오름차순 정렬
5. index를 초기화하고 기존의 인덱스는 삭제

```python
## 1.csv 폴더에 있는 Sales Records.csv 파일 로드

sales = pd.read_csv("../csv/Sales Records.csv")
```

```python
## 2.'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 삭제

# drop_data = ['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel']
# for i in range(0,len(drop_data),1):
#    sales = sales.drop(drop_data[i], axis = 1)
sales = sales.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel'], axis=1)
```

```python
## 3.컬럼의 이름을 변경. ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']

sales.columns = ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
## 4.권역과 국가를 기준으로 오름차순 정렬

# sales = sales.sort_values(by=['권역', '국가'])
# sales = sales.sort_values(['권역', '국가'], inplace=True)
sales = sales.sort_values(['권역', '국가'])
```

```python
## 5.index를 초기화하고 기존의 인덱스는 삭제

# sales.reset_index(drop=True, inplace=True) inplace : 변화시킨 내용을 저장
sales = sales.reset_index(drop=True)
```

```python
sales
```

-  새로운 파생 변수 생성
    1. 총 단가 -> 판매단위*단가
    2. 총원가 -> 판매단위*원가
    3. 총이윤  -> 총단가-총운가

```python
sales['총단가'] = sales['판매단위'] * sales['단가']
sales['총원가'] = sales['판매단위'] * sales['원가']
sales['총이윤'] = sales['총단가'] - sales['총원가']

sales.tail()
```

```python
# 국가 컬럼의 데이터가 항목들의 카운트를 출력하는 함수
sales['국가'].value_counts().index
# sales['국가'].unique()
```

```python
# 국가 하나를 선택해서 그 국가의 총이윤의 합계

sales.loc[sales['국가'] == 'South Korea', '총이윤'] 
# sales.loc[sales['국가'] == 'South Korea']['총이윤']와 다른 방법, 같은 결과 그 이유는?
# 
```

```python
# sales.loc[sales['국가'] == 'South Korea'] : 데이터 프레임
# sales.loc[sales['국가'] == 'South Korea']['총이윤'] : 데이터 프레임에서 해당 컬럼 선택
sales.loc[sales['국가'] == 'South Korea']['총이윤'].sum() # 우리나라의 총 이윤
```

```python
_list = sales['국가'].unique()
profit_list = {}

for i in _list:
    # 총이윤의 합계
    total_profit = sales.loc[sales['국가'] == i]['총이윤'].sum()
    profit_list[i] = total_profit
```

```python
pd.Series(profit_list).sort_values(ascending=False)
```
